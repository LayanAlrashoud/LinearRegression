<h2>  Linear Regression with multiple independent varibles</h2>
<h4>Using linear regression to predict salary based on the test_score(out of 10),interview_score(out of 10) and experience</h4>

```python
import pandas as pd

df = pd.read_csv('hiring.csv')
df
```

```python
med=df['test_score(out of 10)'].median()
```

```python
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(med)
df
```

```python
from word2number import w2n

def convert_text_to_num(text):
    try:                                         
        return w2n.word_to_num(text)
    except ValueError:
        return text 


df['experience'] = df['experience'].apply(convert_text_to_num)

df.to_csv('hiring.csv', index=False)

df
```

```python
med2=df['experience'].median()
df['experience']=df.experience.fillna(med2)
df
```

```python
from sklearn.linear_model import LinearRegression

model2 = LinearRegression()
model2.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])

model2.predict([[11,7,8]])
```
